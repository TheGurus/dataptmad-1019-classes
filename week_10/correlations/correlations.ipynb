```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vehicles = pd.read_csv('../../week_5/pandas_import_export/data/vehicles.csv')
vehicles.head()
```

## Correlation Analysis

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

```python
vehicles.corr()
```

## Plots

### Pandas

```python
plt.scatter(vehicles['City MPG'], vehicles['Highway MPG']);
```

```python
vehicles.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
vehicles.corr(method='spearman').style.background_gradient(cmap='coolwarm').set_precision(2)
```
