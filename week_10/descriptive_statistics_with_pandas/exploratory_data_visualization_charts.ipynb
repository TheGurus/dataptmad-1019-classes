## 1. data loading

```python
import pandas as pd
```

```python
vehicles = pd.read_csv('./data/vehicles.csv')
```

## 2. visualization setup

```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import set_matplotlib_formats

set_matplotlib_formats('svg')

sns.set_style('whitegrid')
```

```python
%matplotlib inline
```

## 3. visualizations

### 3.1 box plot (pandas vs seaborn)

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html

```python
vehicles['Combined MPG'].plot(kind='box');
```

```python
vehicles.boxplot(column=['Combined MPG'], 
                 rot=45,
                 by=['Cylinders'],
                 fontsize=15,
                 figsize=(8, 8));
```

```python
fig = plt.figure(figsize=(8, 8))
ax = sns.boxplot(x='Cylinders', y='Combined MPG', data=vehicles);
ax.set_xticklabels(ax.get_xticklabels(),rotation=45);
```

```python
fig = plt.figure(figsize=(64, 8))
ax = sns.boxplot(hue='Fuel Type', y='Highway MPG', x='Cylinders', data=vehicles, palette='Set3');
ax.set_xticklabels(ax.get_xticklabels(),rotation=45);
```

### 3.2 violin plots (seaborn)

```python
fig = plt.figure(figsize=(8, 8))
ax = sns.violinplot(x='Cylinders', y='Combined MPG', data=vehicles);
```

```python
fig = plt.figure(figsize=(64, 8))
ax = sns.violinplot(x='Fuel Type', 
                    y='Highway MPG', 
                    hue='Cylinders', 
                    data=vehicles, 
                    palette='Set3');
ax.set_xticklabels(ax.get_xticklabels(),rotation=45);
```

### 3.3 histograms (pandas vs seaborn)

https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
vehicles['Fuel Barrels/Year'].hist();
```

```python
vehicles['Fuel Barrels/Year'].hist(bins=50);
```

```python
sns.distplot(vehicles['Fuel Barrels/Year'], bins=10);
```

### 3.4 barplots

```python
vehicles['Cylinders'].value_counts().plot.bar();
```

```python
fig = plt.figure(figsize=(32, 8))
ax = sns.barplot(x='Fuel Type', y='Highway MPG', hue='Cylinders', data=vehicles)
```
